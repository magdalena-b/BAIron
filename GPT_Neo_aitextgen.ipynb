```python
from aitextgen import aitextgen
```

```python
ai = aitextgen(model="EleutherAI/gpt-neo-125M", to_gpu=True)
```

```python
file_name = "shakespeare_sonnets.txt"
```

```python
ai.train(file_name,
         line_by_line=False,
         from_cache=False,
         num_steps=3000,
         generate_every=300,
         save_every=300,
         save_gdrive=False,
         learning_rate=1e-3,
         fp16=False,
         batch_size=1, 
         )
```

```python
ai = aitextgen(model_folder="trained_model", to_gpu=True)
```

```python
ai.generate(n=1,
            batch_size=5,
            prompt="wind has blown the rain away",
            max_length=200,
            temperature=0.7,
            top_p=0.9)
```

```python
ai.generate(n=1,
            batch_size=5,
            prompt="lol",
            max_length=200,
            temperature=1.0,
            top_p=0.9)
```
