```python
from happytransformer import HappyGeneration

happy_gen = HappyGeneration("GPT-NEO", "EleutherAI/gpt-neo-125M")
```

```python
import torch
torch.cuda.empty_cache()
```

```python
from happytransformer import GENTrainArgs 

args = GENTrainArgs(num_train_epochs = 5, batch_size=1)
happy_gen.train("cummings_public_domain.txt", args=args)

# happy_gen.train("cummings_public_domain.txt")
```

```python
result = happy_gen.generate_text("Wind has blown the rain away")
print(result)
print(result.text)
```

```python
happy_gen.save("model/")
```
