```python
from simpletransformers.language_modeling import LanguageModelingModel
import logging


logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)

train_args = {
    "reprocess_input_data": True,
    "overwrite_output_dir": True,
    "train_batch_size": 32,
    "num_train_epochs": 20,
    "mlm": False,
}

model = LanguageModelingModel('gpt2', 'gpt2', args=train_args)

model.train_model(train_file="shakespeare_sonnets_train.txt", output_dir="gpt2-simpletransformers", eval_file="shakespeare_sonnets_test.txt")

model.eval_model("shakespeare_sonnets_test.txt")
```

```python
 from simpletransformers.language_generation import LanguageGenerationModel
 
 finemodel = LanguageGenerationModel(
    model_type = "gpt2", model_name = "gpt2-simpletransformers", args={"max_length": 200}
)
 generated = finemodel.generate("beauty is beginning of terror", verbose=False)
 print(generated)
```
